Conversion rates by origin and landing page

```python
deals_df.columns
```

```python
mql_df.columns
```

```python
# Calculate the number of leads per origin and per landing page
leads_per_origin = mql_df['origin'].value_counts()
leads_per_lp = mql_df['landing_page_id'].value_counts()

# Calculate the number of closed deals per origin and per landing page
deals_per_origin = df['origin'].value_counts()
deals_per_lp = df['landing_page_id'].value_counts()

# Calculate conversion rate per origin and per landing page
crt_per_origin = (deals_per_origin / leads_per_origin) * 100
crt_per_lp = (deals_per_lp / leads_per_lp) * 100
```

```python
crt_per_origin
```

Conversion Rates by Lead Origin:
- Unknown: 16.29%
- Paid Search: 12.30%
- Organic Search: 11.80%
- Direct Traffic: 11.22%
- Referral: 8.45%
- Social: 5.56%
- Email: 3.04%
- Display: 5.08%
- Other Publicities: 4.62%
- Other: 2.67%

From these results, paid search, organic search, and direct traffic have relatively high conversion rates, indicating their strong effectiveness.

### Compare the number of closed deals by business segment

```python
deals_per_business = df['business_segment'].value_counts()

plt.figure(figsize=(12, 6))
deals_per_business.plot(kind='bar', color='lightcoral')
plt.title('# Closed Deals per Business Segment')
plt.xlabel('Business Segment')
plt.ylabel('# Deals')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
deals_per_business = df['business_segment'].value_counts().reset_index()
deals_per_business['percentage'] = ((deals_per_business['count'] / deals_per_business['count'].sum()) * 100).round(2)
```

```python
deals_per_business.head(10)
```

```python
origin_by_business_segment = df.groupby(['business_segment', 'origin']).size().unstack(fill_value=0)
```

```python
origin_by_business_segment
```

```python
deals_per_business = df['business_segment'].value_counts()

# Plot a stacked bar chart for lead origins by business segment (top 10 business segments)
top_10_business_segments = deals_per_business.head(10).index
origin_by_business_segment_top10 = origin_by_business_segment.loc[top_10_business_segments]


plt.figure(figsize=(12, 6))
origin_by_business_segment_top10.plot(kind='bar', stacked=True)
plt.title('Lead Origins by Business Segment (Top 10)')
plt.xlabel('Business Segment')
plt.ylabel('Number of Deals')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Lead Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Lead Origins by Business Segment (Top 10): The stacked bar chart highlights how different lead origins contribute to deals within the top 10 business segments. For instance, organic search and paid search are significant contributors to business segments like home appliances and car accessories, while referral and social also play notable roles in certain segments.
