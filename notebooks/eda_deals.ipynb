```python
from scipy.stats import chi2_contingency
import itertools
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2, p, dof, ex = chi2_contingency(confusion_matrix)
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))  # Correction for bias
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))


# Ensure categorical variables are of type 'category'
categorical_cols = deals_df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    deals_df[col] = deals_df[col].astype('category')

categorical_vars = deals_df[categorical_cols].columns
pairs = list(itertools.combinations(categorical_vars, 2))

# Initialize the association matrix
association_matrix = pd.DataFrame(np.zeros((len(categorical_vars), len(categorical_vars))),
                                  index=categorical_vars,
                                  columns=categorical_vars)

# Compute Cramer's V for each pair
for var1, var2 in pairs:
    association = cramers_v(deals_df[var1], deals_df[var2])
    association_matrix.loc[var1, var2] = association
    association_matrix.loc[var2, var1] = association

# Set diagonal to 1
np.fill_diagonal(association_matrix.values, 1)

print(association_matrix)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(association_matrix, annot=True, cmap='coolwarm')
plt.title("Cramer's V Association Matrix")
plt.show()
```
